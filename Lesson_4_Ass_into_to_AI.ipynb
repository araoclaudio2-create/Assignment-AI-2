```python
import pandas as pd
import numpy as np
from functools import reduce

# Task 1: Create a DataFrame using Pandas
# Create a DataFrame with the following data:
# Name: Alice, Bob, Charlie, David
# Age: 25, 30, 35, 28
# City: New York, San Francisco, Los Angeles, Chicago

# Your code here:
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Toronto', 'Montreal']
}

df = pd.DataFrame(data)


# Task 2: Row and Column Manipulation
# Drop the 'City' column from the DataFrame created in Task 1

# Your code here:
df_dropped = df.drop(columns=['City'], errors='ignore')

print("\nDataFrame after dropping 'City':\n", df_dropped)

# Task 3: Handling Null Values
# Create a new DataFrame with some null values and fill them

# Your code here:
# Create DataFrame with null values
df_null = pd.DataFrame({
    'A': [1, np.nan, 3],
    'B': [np.nan, 5, 6],
    'C': [7, 8, np.nan]
})

print("\nDataFrame with nulls:\n", df_null)

# Fill null values
df_filled = df_null.fillna(0)

print("\nAfter filling nulls with 0:\n", df_filled)

# other methods
df_dropna = df_null.dropna()
print("\nAfter dropping null rows:\n", df_dropna)

# Task 4: GroupBy and Describe
# Using the following DataFrame, group by 'Category' and describe the 'Value' column

df_group = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'B', 'A', 'C'],
    'Value': [10, 20, 15, 25, 30, 35]
})

# Your code here:
grouped = df_group.groupby('Category')['Value'].describe()

print("\nGroupBy Describe:\n", grouped)

# Interpretation:
print("\nInterpretation: Shows count, mean, std, min, max for each category.")

# Task 5: Concatenation and Merging
# Concatenate two DataFrames vertically and merge them horizontally

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
df3 = pd.DataFrame({'C': [9, 10], 'D': [11, 12]})

# Your code here:
concat_df = pd.concat([df1, df2], axis=0).reset_index(drop=True)

# Merge horizontally
merged_df = pd.concat([concat_df, df3], axis=1)

print("\nConcatenated DataFrame:\n", concat_df)
print("\nMerged DataFrame:\n", merged_df)
print("\nShape:", merged_df.shape)

# Task 6: Tuples and Sets
# Create a tuple of fruits and a set of numbers.
# Then, try to add an element to each and observe the difference.

# Your code here:
# Tuple
fruits = ('apple', 'banana', 'cherry')

# Set
numbers = {1, 2, 3, 4, 5}

# Attempt to add
try:
    fruits[0] = 'orange'  # Will fail
except TypeError as e:
    print("\nTuple Error:", e)

numbers.add(6)

print("\nSet after adding:", numbers)

print("\nExplanation: Tuples are immutable, sets are mutable.")

# Task 7: Dictionaries
# Create a dictionary of student names and their scores.
# Then, update a student's score and add a new student.

# Your code here:
students = {'Alice': 85, 'Bob': 90, 'Charlie': 78}

# Update score
students['Alice'] = 88

# Add new student
students['David'] = 92

print("\nUpdated Dictionary:\n", students)

# Task 8: Functions and Lambda
# Create a function to calculate the square of a number.
# Then, use a lambda function to do the same.

# Your code here:
# Regular function
def square(x):
    return x ** 2

# Lambda function
square_lambda = lambda x: x ** 2

print("\nRegular Function:", square(4), square(5))
print("Lambda Function:", square_lambda(4), square_lambda(5))

# Task 9: Iterators and Generators
# Create an iterator for the first 5 even numbers.
# Then, create a generator for the same.

# Your code here:
# Iterator class
class EvenNumbers:
    def __init__(self):
        self.num = 0
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < 5:
            result = self.num
            self.num += 2
            self.count += 1
            return result
        else:
            raise StopIteration

print("\nIterator Output:")
for num in EvenNumbers():
    print(num)

# Generator
def even_generator():
    num = 0
    for _ in range(5):
        yield num
        num += 2

print("\nGenerator Output:")
for num in even_generator():
    print(num)

# Task 10: Map, Reduce, and Filter
# Use map to square all numbers in a list.
# Use reduce to find the product of all numbers in a list.
# Use filter to get only even numbers from a list.

numbers = [1, 2, 3, 4, 5]

# Your code here:
from functools import reduce

nums = [1, 2, 3, 4, 5]

# Map
squared = list(map(lambda x: x**2, nums))

# Reduce
product = reduce(lambda x, y: x * y, nums)

# Filter
even = list(filter(lambda x: x % 2 == 0, nums))

print("\nSquared:", squared)
print("Product:", product)
print("Even Numbers:", even)

# Task 11: Object-Oriented Programming - Creating a Class
# Create a class called 'Rectangle' with attributes 'length' and 'width'.
# Include methods to calculate area and perimeter.

# Your code here:
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

# Create objects
rect1 = Rectangle(4, 5)
rect2 = Rectangle(6, 7)

print("\nRectangle 1 Area:", rect1.area())
print("Rectangle 1 Perimeter:", rect1.perimeter())

print("\nRectangle 2 Area:", rect2.area())
print("Rectangle 2 Perimeter:", rect2.perimeter())

# Task 12: Pandas Data Analysis
# Using the following DataFrame, perform these tasks:
# 1. Find the average salary by department
# 2. Get the names of employees with salary > 60000
# 3. Add a new column 'Bonus' which is 10% of salary

df_employees = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [55000, 65000, 70000, 60000, 58000]
})

# Your code here:
# Average salary by department
avg_salary = df_employees.groupby('Department')['Salary'].mean()
print("\nAverage Salary:\n", avg_salary)

# Employees with salary > 60000
high_salary = df_employees[df_employees['Salary'] > 60000]['Name']
print("\nEmployees with salary > 60000:\n", high_salary)

# Add Bonus column
df_employees['Bonus'] = df_employees['Salary'] * 0.10

print("\nFinal DataFrame with Bonus:\n", df_employees)
```
