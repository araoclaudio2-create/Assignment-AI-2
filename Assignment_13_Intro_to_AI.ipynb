```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import TextVectorization, Embedding, LSTM, Dense
from tensorflow.keras.models import Sequential

# 2. Load Dataset
text = """This is a simple sample dataset. You can replace this with a larger dataset from Project Gutenberg."""

# 3. Preprocess Text
vectorizer = TextVectorization(output_mode='int', output_sequence_length=20)
vectorizer.adapt([text])
vocab_size = len(vectorizer.get_vocabulary())

# 4. Create Sequences
sequences = vectorizer([text])
X = sequences[:, :-1]
y = sequences[:, 1:]

# 5. Build Model
model = Sequential([
    Embedding(vocab_size, 64), # input_length is no longer needed in modern Keras versions
    LSTM(128, return_sequences=True),
    Dense(vocab_size, activation='softmax')
])
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')
model.summary()

# 6. Train Model
model.fit(X, y, epochs=10)

# 7. Text Generation Function
def generate_text(seed, num_words=20):
    for _ in range(num_words):
        tokenized = vectorizer([seed])
        prediction = model.predict(tokenized)
        next_word = np.argmax(prediction[0, -1])
        word = vectorizer.get_vocabulary()[next_word]
        seed += " " + word
    return seed

print(generate_text("This is"))
```
